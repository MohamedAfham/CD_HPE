```python
with open('valid_cover2.json', 'r') as f:
    data = json.load(f)
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import cv2

import torchvision.transforms.functional as F

plt.rcParams["savefig.bbox"] = 'tight'

def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fix, axs = plt.subplots(ncols=len(imgs), squeeze=False)
    for i, img in enumerate(imgs):
        axs[0, i].imshow(np.asarray(img))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])


def draw_keypoints(keypoints, image):
    # the `outputs` is list which in-turn contains the dictionaries 
    for i in range(len(keypoints)):
        keypoint = keypoints[i]
        cv2.circle(image, (int(keypoint[0]), int(keypoint[1])), 1, (1 ,1, 0), thickness=3)
    return image
```

```python
id = 25
im = plt.imread(data[id]['file_name'])
plt.imshow(im)

keypoints = np.array(data[id]['key_points'])[:, :2]

result = draw_keypoints(keypoints, im)
show(result)
```
